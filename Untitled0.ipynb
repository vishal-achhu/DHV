```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,3,4,5]
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Plot")
plt.show
```

```python
import matplotlib.pyplot as py
x=[1,2,3,4,5]
y=[2,4,5,6,7]
colors=["r","y","b","g","b"]
sizes=[100,200,300,40,15]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,marker="*",edgecolor="g")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.ylabel("y axis",fontsize=20)
plt.show()
```

```python
x1=[3,3,4,5,6]
plt.hist(x1,color="b",edgecolor="g",bottom=1,histtype="bar",orientation="horizontal",label="python")
plt.axvline(5,color="g")
plt.legend()
plt.show()
```

```python
import numpy as np
y=np.array([35,25,25,15])
mylabels=["banana","apple","cherries","dates"]
myexplode=[0.2,0.4,0.8,0.9]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```
