```python
import pandas as pd

```

```python

data = {
    'Age_group': ['<21', '21-24','24-27','27-30','30-33','33-36','36-39','57-60','>60'],
    'T_loans': [9,310,511,4000,4568,5698,8209,788,6],
    'N_badloans': [2,14,20,172,169,188,197,9,0],
    'N_goodloans': [7,296,491,3828,4399,5510,8012,779,6],
    'P_badloans': ['22%','4.5%','3.9%','4.3%','3.7%','3.3%','2.4%','1.1%','0.0%'],
    'P_goodloans': ['77.8%','95.5%','96.1%','95.7%','96.3%','96.7%','97.6%','98.9%','100%']
}

df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt


plt.bar(df['Age_group'], df['T_loans'])
plt.xlabel('Age_group')
plt.ylabel('T_loans')
plt.title('Bar Graph on Age_group vs total number of loans')
plt.show()
```

```python
plt.plot(df['Age_group'], df['T_loans'])
plt.xlabel('Age group')
plt.ylabel('Total Loans')
plt.title('Line Plot Age_group vs total number of loans')
plt.show()
```

```python

plt.scatter(df['Age_group'], df['T_loans'])
plt.xlabel('Age group')
plt.ylabel('Total loans')
plt.title('Scatter Plot of Age group vs total loans')
plt.show()
```

```python
plt.bar(df['N_badloans'], df['N_goodloans'])
plt.xlabel('Bad loans')
plt.ylabel('Good loans')
plt.title('Bar Graph on Good loans vs Bad loans')
plt.show()
```

```python
plt.scatter(df['N_badloans'], df['N_goodloans'])
plt.xlabel('Bad loans')
plt.ylabel('Good loans')
plt.title('scatter plot on Good loans vs Bad loans')
plt.show()
```

```python
plt.plot(df['Age_group'], df['T_loans'])
plt.xlabel('Age group')
plt.ylabel('Total loans')
plt.title('line plot Plot of Age group vs total loans')
plt.show()
```

```python
plt.bar(df['P_badloans'], df['P_goodloans'])
plt.xlabel('Percentage of bad loans')
plt.ylabel('Percentage of good loans')
plt.title('Bar Graph on Age_group vs total number of loans')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(df['T_loans'], df['N_badloans'], df['N_goodloans'])

ax.set_xlabel('total no of loans')
ax.set_ylabel('Number of bad loans')
ax.set_zlabel('Number of good loans')

plt.title('3D Scatter Plot of Total loans vs Bad loans vs Good loans')
plt.show()
```
