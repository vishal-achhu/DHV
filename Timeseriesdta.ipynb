```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data={
    "date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01"]),
    "value":[10,15,22,28,35]
}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='o',color="red")
plt.xlabel('Date')
plt.ylabel("Value")
plt.title("Time series with trend")
plt.grid(True)
plt.show()
```
